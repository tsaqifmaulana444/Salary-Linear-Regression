```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Salary_dataset.csv")

years_experience = df['YearsExperience']
salary = df['Salary']
```

```python
plt.scatter(years_experience, salary)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')

regressor = LinearRegression()
regressor.fit(years_experience.values.reshape(-1, 1), salary)

line = regressor.coef_ * years_experience + regressor.intercept_
plt.plot(years_experience, line, color='red')

plt.show()
```

The math formula for a linear regression line can be represented as:

y = m * x + b

Where:

    y is the dependent variable (in this case, "Salary")
    x is the independent variable (in this case, "YearsExperience")
    m is the slope of the line (regression coefficient)
    b is the y-intercept of the line (regression intercept)
